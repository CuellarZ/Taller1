```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
promedio_edad = df['Age'].mean()
```

```python
valores_nulos_por_columna = df.isnull().sum()
```

```python
genero_mas_repetido =df['Gender'].mode()
```

```python
hombres = (df['Gender']==0).sum()
hombres
```

```python
mujeres = (df['Gender']==1).sum()
mujeres
```

```python
Sports = (df['Sports']==1).sum()
Sports
```

```python
estudiantes_deportes_por_genero = df.groupby('Gender')['Sports'].sum()
print(estudiantes_deportes_por_genero)
```

```python
hombres_que_practican_deportes = df[(df['Gender'] == 0) & (df['Sports'] == 1)]['Sports'].count()
mujeres_que_practican_deportes = df[(df['Gender'] == 1) & (df['Sports'] == 1)]['Sports'].count()

print(f"Hombres que practican deportes: {hombres_que_practican_deportes}")
print(f"Mujeres que practican deportes: {mujeres_que_practican_deportes}")
```

```python
tutorias = (df['Tutoring']==1).sum()
tutorias
```

```python
tiempo_estudio_maximo = df['StudyTimeWeekly'].max()
tiempo_estudio_minimo = df['StudyTimeWeekly'].min()

print(f"Tiempo de estudio máximo por semana: {tiempo_estudio_maximo}")
print(f"Tiempo de estudio mínimo por semana: {tiempo_estudio_minimo}")
```
